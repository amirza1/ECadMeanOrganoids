```python
import numpy as np
import time, os, sys
from urllib.parse import urlparse
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rcParams['figure.dpi'] = 200
import random
import PIL
import pandas as pd
```

```python
analysis_dir = '/path/to/dir'
```

```python
control_features = pd.read_csv(analysis_dir + '/control_features.csv')
treatment_features = pd.read_csv(analysis_dir + '/treatment_features.csv')
```

```python
# Add a column to identify control and treatment
control_features['group'] = 'control'
treatment_features['group'] = 'treatment'

# Concatenate the two dataframes
all_features = pd.concat([control_features, treatment_features])
all_features['patient'] = [x.split('_')[0] for x in all_features['name']]
all_features['treatment'] = [x.split('_')[2] for x in all_features['name']]
all_features['series'] = [x.split('_')[3] for x in all_features['name']]
```

```python
all_features.groupby(['patient', 'treatment']).count()
```

```python
patient2select = {
    'BR63T': 146,
    'BR72T': 73,
    'BR73T': 385,
    'BR83T': 49
}
```

```python
sampled_features = all_features.groupby(['patient', 'treatment']).apply(lambda x: x.sample(n=patient2select[x.patient.values[0]], random_state=1)).reset_index(drop=True)
sampled_features.to_csv(analysis_dir + '/sampled_features.csv', index=False)
```
