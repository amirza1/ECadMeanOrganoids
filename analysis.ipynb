```python
analysis_dir = '/path/to/dir'
control_features = pd.read_csv(analysis_dir + '/control_features.csv')
treatment_features = pd.read_csvanalysis_dir + '/treatment_features.csv')
```

```python
# Add a column to identify control and treatment
control_features['group'] = 'control'
treatment_features['group'] = 'treatment'

# Concatenate the two dataframes
all_features = pd.concat([control_features, treatment_features])
all_features['patient'] = [x.split('_')[0] for x in all_features['name']]
all_features['treatment'] = [x.split('_')[2] for x in all_features['name']]
all_features['series'] = [x.split('_')[3] for x in all_features['name']]
```
